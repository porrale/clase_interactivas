## 1. El Modelo Matemático

Analizaremos una señal senoidal pura definida por la ecuación fundamental de tensión en el tiempo:

$$V(t) = V_p \cdot \sin(2\pi \cdot f \cdot t)$$

Donde:
- $V_p$: Amplitud máxima o pico [Voltios]
- $f$: Frecuencia de oscilación [Hertz]
- $t$: Vector de tiempo [Segundos]

```python
# Importación de librerías esenciales para cómputo científico y gráficos
!pip install numpy
!pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Configuración para renderizar gráficos dentro del flujo de la presentación
%matplotlib inline
```

## 2. Simulación de la Señal y Gráfico en Vivo

Modificando las variables en la celda de abajo, podemos alterar la frecuencia y la amplitud directamente durante la exposición para observar el comportamiento dinámico de la señal.

```python
# --- PARÁMETROS MODIFICABLES EN VIVO ---
frecuencia = 50   # Frecuencia en Hz (Probá cambiar a 100 o 150)
amplitud = 5      # Amplitud en Voltios
# ---------------------------------------

# Generación del vector de tiempo (0 a 40 milisegundos para ver un par de ciclos a 50Hz)
t = np.linspace(0, 0.04, 1000)
v = amplitud * np.sin(2 * np.pi * frecuencia * t)

# Construcción del gráfico
plt.figure(figsize=(10, 4.5))
plt.plot(t * 1000, v, color='#1f77b4', linewidth=2.5, label=f'Señal de {frecuencia} Hz')
plt.title("Visualización de Tensión en Función del Tiempo", fontsize=14, fontweight='bold')
plt.xlabel("Tiempo [ms]", fontsize=12)
plt.ylabel("Tensión [V]", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend(loc='upper right', fontsize=11)
plt.ylim(-amplitud - 2, amplitud + 2)
plt.show()
```
